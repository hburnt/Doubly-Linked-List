```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
url = 'https://raw.githubusercontent.com/hburnt/Intro-to-ML-datasets/main/D3.csv'

df = pd.read_csv(url)
```

```python
#Separating input values and output values
X1 = df.values[:,0]
X2 = df.values[:,1]
X3 = df.values[:,2]
m = len(X1)
Y = df.values[:,3]
```

```python
#Gradient Descent for X1:
def gradient_descent(theta0_now, theta1_now, X, Y, learningRate):
  theta0_gradient = 0
  theta1_gradient = 0

  m = len(X)

  for i in range(m):
    x = X[i]
    y = Y[i]

    theta0_gradient += -(1/m) * ( y - ( theta1_now * x + theta0_now ) )
    theta1_gradient += -(1/m) * x * ( y - ( theta1_now * x + theta0_now ) )

  theta0 = theta0_now - learningRate * theta0_gradient
  theta1 = theta1_now - learningRate * theta1_gradient

  return theta0, theta1
```

```python
def loss_function(theta0, theta1, X, Y):
  total_error = 0
  for i in range(len(Y)):
    x = X[i]
    y = Y[i]
    total_error += (y - (theta1 * x + theta0)) ** 2
  return total_error / float(len(Y))
```

```python
learningRate = 0.05
epochs = 1500

#Initializing theta values for X1, X2, and X3
theta0_X1 = 0
theta1_X1 = 0

theta0_X2 = 0
theta1_X2 = 0

theta0_X3 = 0
theta1_X3 = 0

# Initialize arrays to store the loss values for X1, X2, and X3
cost1 = np.zeros(epochs)
cost2 = np.zeros(epochs)
cost3 = np.zeros(epochs)

Iterations = range(1, epochs+1)

for i in range(epochs):
  # Perform Gradient Descent and record loss for X1, X2, and X3
  theta0_X1, theta1_X1 = gradient_descent(theta0_X1, theta1_X1, X1, Y, learningRate)
  theta0_X2, theta1_X2 = gradient_descent(theta0_X2, theta1_X2, X2, Y, learningRate)
  theta0_X3, theta1_X3 = gradient_descent(theta0_X3, theta1_X3, X3, Y, learningRate)
  cost1[i] = loss_function(theta0_X1, theta1_X1, X1, Y)
  cost2[i] = loss_function(theta0_X2, theta1_X2, X2, Y)
  cost3[i] = loss_function(theta0_X3, theta1_X3, X3, Y)

# Create subplots for X1, X2, and X3
plt.subplots(3,2,figsize=(12,10))
plt.tight_layout(pad=4)

# Regression plot for X1
plt.subplot(3,2,1)
plt.scatter(X1, Y, color="red", marker = '+', label='Training Data')
plt.plot(list(range(0, 5 )), [theta1_X1 * X1 + theta0_X1 for X1 in range(0, 5)], color="blue")

plt.grid(True)
plt.legend()
plt.title('Linear Regression Model for X1')
plt.xlabel('Input')
plt.ylabel('Output')

# Gradient descent plot for X1
plt.subplot(3,2,2)

plt.plot(Iterations, cost1, color='green')

plt.grid(True)
plt.title('Gradient Descent for Theta 1')
plt.xlabel('Iterations')
plt.ylabel('Loss')

##------------------------------------------------------------------------------##

#Regression plot for X2
plt.subplot(3,2,3)
plt.scatter(X2, Y, color="red", marker = '+', label='Training Data')
plt.plot(list(range(0, 5 )), [theta1_X2 * X2 + theta0_X2 for X2 in range(0, 5)], color="blue")

plt.grid(True)
plt.legend()
plt.title('Linear Regression Model for X2')
plt.xlabel('Input')
plt.ylabel('Output')

# Gradient descent plot for X2
plt.subplot(3,2,4)
plt.plot(Iterations, cost2, color='green')

plt.grid(True)
plt.title('Gradient Descent for Theta 2')
plt.xlabel('Iterations')
plt.ylabel('Loss')

##------------------------------------------------------------------------------##

#Regression plot for X3
plt.subplot(3,2,5)
plt.scatter(X3, Y, color="red", marker = '+', label='Training Data')
plt.plot(list(range(0, 5 )), [theta1_X3 * X3 + theta0_X3 for X3 in range(0, 5)], color="blue")

plt.grid(True)
plt.legend()
plt.title('Linear Regression Model for X3')
plt.xlabel('Input')
plt.ylabel('Output')

# Gradient descent plot for X3
plt.subplot(3,2,6)
plt.plot(Iterations, cost3, color='green')

plt.grid(True)
plt.title('Gradient Descent for Theta 3')
plt.xlabel('Iterations')
plt.ylabel('Loss')
```
